```python
# Created by: Francisco Rodríguez Candón
# Last update: 25/10/2024
# mail: francandon@unizar.es

import os
import sys
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from FluxResponse.FOVPlotter import FOVPlotter

# Update matplotlib parameters for styling
mpl.rcParams.update({'font.size': 24, 'font.family': 'serif'})
# mpl.rcParams.update({'font.size': 18, 'font.serif': 'Palatino'})
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1
mpl.rcParams['xtick.minor.size'] = 3
mpl.rcParams['xtick.minor.width'] = 1
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1
mpl.rcParams['ytick.minor.size'] = 3
mpl.rcParams['ytick.minor.width'] = 1
mpl.rc('text', usetex=True)

mpl.rcParams['xtick.direction'] = 'in'
mpl.rcParams['ytick.direction'] = 'in'
mpl.rcParams['xtick.top'] = True
mpl.rcParams['ytick.right'] = True

# Define your custom colors
blue, red, green, purple, orange, magenta = '#39558c', '#d35d47', '#308f69', '#71338c', '#d38e47', '#ac3a76'
golden, turquoise, lavender, coral, lime, fuchsia, indigo, amber, teal = '#f5a953', '#5ca8b1', '#8e6385', '#c74d36', '#6ca95c', '#be519d', '#5e79b8', '#d18536', '#4cbfad'

# Create a custom colormap using one of your colors
from matplotlib.colors import LinearSegmentedColormap

def create_colormap_from_color(hex_color, hex_color_2,hex_color_3, hex_color_4, name='custom_cmap'):
    import matplotlib.colors as mcolors
    cmap = LinearSegmentedColormap.from_list(name, ['white',hex_color_2, hex_color_4, red])
    return cmap

# Choose a color to create the colormap
custom_cmap = create_colormap_from_color(indigo, blue, orange, coral, name='blue_cmap')

# Get the current working directory
current_dir = os.getcwd()

# Determine the root directory (one level up from the current directory)
root_dir = os.path.abspath(os.path.join(current_dir, '..'))

# Append the root directory to sys.path
sys.path.append(root_dir)
FOV_dir = os.path.join(root_dir,'Python-Events-File-Plotter', 'Observations')
PLOT_dir = os.path.join(root_dir,'Python-Events-File-Plotter', 'Plots')

# Initialize the FOVPlotter with the path to your FITS file
fits_path = os.path.join(FOV_dir, 'nu30501012002A01_cl.evt')
plotter = FOVPlotter(fits_path, num_bins=200, cmap="CMRmap", log_scale=True, fov_size=0.21, legend_fontsize=15)

# Adds source circle
plotter.add_circle(
    ra_center=88.7929583,     # Replace with your RA center in degrees
    dec_center=7.4070556,      # Replace with your Dec center in degrees
    radius=60/3600,         # Replace with your desired radius in degrees
    color='crimson',
    lw=2,
    linestyle='-',
    label='Source Region'
)

# Adds background circle
plotter.add_circle(
    ra_center=88.7716098,     # Replace with your RA center in degrees
    dec_center=7.3482892,      # Replace with your Dec center in degrees
    radius=120/3600,         # Replace with your desired radius in degrees
    color='#a6d96a',
    lw=2,
    linestyle='--',
    label='Circular Background'
)

# Adds the polygon
# Define the coordinates of the polygon as a flat list
polygon_coords = [
    88.8230167, 7.3972246,
    88.8080983, 7.3773098,
    88.7880171, 7.3704814,
    88.7696561, 7.3841365,
    88.7612763, 7.4124722,
    88.7157232, 7.3795770,
    88.7673670, 7.3152867,
    88.8419494, 7.3716182
]
plotter.add_polygon(
    coords=polygon_coords,
    color='#fee08b',
    lw=2,
    linestyle='--',
    label='Polygon Background'
)

# Display and save the plot
plotter.show()
plotter.save(os.path.join(PLOT_dir, 'nu30501012002A01_cl.evt.pdf'))
```
